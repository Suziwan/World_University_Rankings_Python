c) How many universities are taken into account annually by each ranking on average? In other terms, give the average number of universities taken into account by each ranking during a year.

```python
# For timesData
times_avg_universities = timesData.groupby('year')['world_rank'].count().mean()
print("The average number of universities taken into account each year by timesData is:", times_avg_universities)

# For shanghaiData
shanghai_avg_universities = shanghaiData.groupby('year')['world_rank'].count().mean()
print("The average number of universities taken into account each year by shanghaiData is:", shanghai_avg_universities)

# For cwurData
cwur_avg_universities = cwurData.groupby('year')['world_rank'].count().mean()
print("The average number of universities taken into account each year by cwurData is:", cwur_avg_universities)
```

d) What are the rankings obtained by the University of Strasbourg in each of the 3 available rankings?

```python
# For timesData
times_uni_strasbourg = timesData[timesData['university_name'] == 'University of Strasbourg']
if not times_uni_strasbourg.empty:
    print("University of Strasbourg in timesData:")
    print(times_uni_strasbourg[['year', 'world_rank']])
    print("\n")
else:
    print("University of Strasbourg is not present in timesData.\n")

# For shanghaiData
shanghai_uni_strasbourg = shanghaiData[shanghaiData['university_name'] == 'University of Strasbourg']
if not shanghai_uni_strasbourg.empty:
    print("University of Strasbourg in shanghaiData:")
    print(shanghai_uni_strasbourg[['year', 'world_rank']])
    print("\n")
else:
    print("University of Strasbourg is not present in shanghaiData.\n")

# For cwurData
cwur_uni_strasbourg = cwurData[cwurData['institution'] == 'University of Strasbourg']
if not cwur_uni_strasbourg.empty:
    print("University of Strasbourg in cwurData:")
    print(cwur_uni_strasbourg[['year', 'world_rank']])
    print("\n")
else:
    print("University of Strasbourg is not present in cwurData.\n")
```

e) According to the Times ranking, what are the top 20 universities in 2016? What is the share of American universities?

```python
# Top 20 universities from 2016 in Times ranking
timesData_2016 = timesData[timesData['year'] == 2016]
top_20_times = timesData_2016.sort_values(by='world_rank').head(20)
print(top_20_times[['university_name', 'world_rank']])

# Percentage of top 20 universities that are in the USA
num_usa_universities = top_20_times[top_20_times['country'] == 'United States of America']['country'].count()
percent_usa_universities = (num_usa_universities / 20) * 100
print("Percentage of top 20 universities in 2016 in the USA: ", percent_usa_universities)
```

f) How many universities does each country have in the Shanghai ranking in 2015? Arrange the list in descending order. Remember to use the university - country table to carry out this request.

```python
# Merge the shanghaiData and school_and_country_table dataframes
merged_data = pd.merge(shanghaiData, school_and_country_table, left_on='university_name', right_on='school_name')

# Filter the data to include only 2015
shanghai_2015 = merged_data[merged_data['year'] == 2015]

# Count the number of universities per country
country_counts = shanghai_2015['country'].value_counts()

# Print the results in descending order
print(country_counts.sort_values(ascending=False))
```

h) Which countries have the lowest ratios (Expenditure on Higher Education Institutions (Private and Public) in 2011) / Number of universities in the Times ranking in 2011? Arrange this list in ascending order.

```python
# Filter education expenditure data for 2011 and higher education institutions only
edu_exp_2011 = education_expenditure[['country', 'institute_type', '2011']]
edu_exp_2011 = edu_exp_2011[(edu_exp_2011['institute_type'].str.strip() == 'Higher Education Institutions')]

# Group by country and sum expenditure
exp_by_country = edu_exp_2011.groupby('country').sum(numeric_only=True)

# Merge with Times ranking data and count number of universities
times_2011 = timesData[timesData['year'] == 2011][['world_rank', 'country']]
uni_count = times_2011.groupby('country').size().reset_index(name='uni_count')
exp_ratio = pd.merge(exp_by_country, uni_count, on='country')
exp_ratio['ratio'] = exp_ratio['2011'] / exp_ratio['uni_count']

# Sort by ratio in ascending order
exp_ratio_sorted = exp_ratio[['country', 'ratio']].sort_values(by='ratio', ascending=True)

exp_ratio_sorted
```

### 3. Find interesting information

Give some extra interesting conclusions from your analysis:

a) Top university each year according to each ranking

```python
# Filter the data for Times ranking
timesData = timesData[timesData['world_rank'].notna()]

# Convert year column to string type and extract the year
timesData['world_rank_numeric'] = pd.to_numeric(timesData['world_rank'], errors='coerce')

# Find the best university per year according to Times ranking
best_universities_times = timesData.loc[timesData.groupby('year')['world_rank_numeric'].idxmin()][['year', 'university_name']]
best_universities_times
```

```python
# Filter the data for Shanghai ranking
shanghaiData = shanghaiData[shanghaiData['world_rank'].notna()]

# Convert year column to string type and extract the year
shanghaiData['world_rank_numeric'] = pd.to_numeric(shanghaiData['world_rank'], errors='coerce')

# Find the best university per year according to Times ranking
best_universities_shanghai = shanghaiData.loc[shanghaiData.groupby('year')['world_rank_numeric'].idxmin()][['year', 'university_name']]
best_universities_shanghai
```

```python
# Filter the data for Shanghai ranking
cwurData = cwurData[cwurData['world_rank'].notna()]

# Convert year column to string type and extract the year
cwurData['world_rank_numeric'] = pd.to_numeric(cwurData['world_rank'], errors='coerce')

# Find the best university per year according to Times ranking
best_universities_cwur = cwurData.loc[cwurData.groupby('year')['world_rank_numeric'].idxmin()][['year', 'institution']]
best_universities_cwur
```

b) Gender equality: Universities having more than 45% of women (Times ranking)

```python
# Replace '-' with NaN in 'female_male_ratio' column
timesData['female_male_ratio'] = timesData['female_male_ratio'].replace('-', np.nan)

# Convert 'female_male_ratio' to percentage of females
timesData['female_percent'] = pd.to_numeric(timesData['female_male_ratio'].str.split(':').str[0]) * 100 / (pd.to_numeric(timesData['female_male_ratio'].str.split(':').str[0]) + pd.to_numeric(timesData['female_male_ratio'].str.split(':').str[1]))

# Filter out universities with less than 45% female students
timesData = timesData[timesData['female_percent'] >= 45]

# Display university name and female_percent columns
print(timesData[['world_rank', 'university_name', 'female_percent', 'year']])
```

c) International students: Universities having the most international students (Times ranking)

```python
# Filter the necessary columns and rows for international students
times_intl = timesData[['world_rank', 'university_name', 'country', 'international_students']]
times_intl = times_intl.dropna() # Drop any rows with missing data

# Convert percentage strings to floats
times_intl['international_students'] = times_intl['international_students'].str.replace('%', '').astype(float)

# Sort by international students in descending order
times_intl_sorted = times_intl.sort_values(by='international_students', ascending=False)

# Display top 10 universities by international students
times_intl_sorted.head(5)[['world_rank', 'university_name', 'country', 'international_students']]
```
